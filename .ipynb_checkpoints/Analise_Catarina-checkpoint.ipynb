```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('C:/Users/roc1lis/Desktop/HR_DS.csv')
#para ver atualizacoes no git
#nova atualização
```

```python
data.head(5)
```

```python
#Checking the Null Values

data.isnull().sum()
```

```python
plt.subplots(figsize=(15,5))
data['Department'].value_counts(normalize = True)
data['Department'].value_counts(dropna = False).plot.bar(color=['black', 'red', 'green', 'blue', 'cyan'])
plt.show()
```

```python
fig = plt.figure(figsize=(20,5))
sns.countplot(x='BusinessTravel', data=data, hue='Department');
```

```python
fig = plt.figure(figsize=(20,5))
sns.countplot(x='JobRole', data=data, hue='MaritalStatus');
```

```python
data['Department'].value_counts()
```

```python
df = data.groupby('EducationField').size()

# Make the plot with pandas
df.plot(kind='pie', subplots=True, figsize=(15, 8))
plt.title("Pie Chart of different types of education")
plt.ylabel("")
plt.show()
```

```python
plt.subplots(figsize=(15,5))
sns.countplot(x = 'EducationField', data = data, hue = 'Gender', palette = 'dark')
plt.show()
```

```python
plt.subplots(figsize=(15,5))
sns.distplot(data['Age'])
plt.title('Distribution of Age of Employees', fontsize = 30)
```

```python
# checking the distribution of length of service
plt.subplots(figsize=(15,8))
sns.distplot(data['YearsAtCompany'], color = 'green')
plt.title('Distribution of length of service among the Employees', fontsize = 30)
plt.xlabel('Length of Service in years')
plt.ylabel('count')
plt.show()
```
